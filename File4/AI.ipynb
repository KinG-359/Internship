```python
!pip install bs4
!pip install requests

from bs4 import BeautifulSoup
import requests

Scrap=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
Scrap

soup=BeautifulSoup(Scrap.content)

```

```python
Titles=soup.find('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg")
Titles.text
```

```python
Titles=[]
for i in soup.find_all('h1',class_="sc-1qrq3sd-0 sc-orwwe2-4 iUciVs fuwoaH"):
    Titles.append(i.text)
Titles
```

```python
Paper_Heading=soup.find('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg")
Paper_Heading.text
   
```

```python
Paper_Heading=[]

for i in soup.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    Paper_Heading.append(i.text)
    
Paper_Heading
```

```python
Writer=soup.find('span',class_="sc-1w3fpd7-0 dnCnAO")
Writer.text

    
```

```python
Writer=[]

for i in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    Writer.append(i.text.split('|')[0])
    
Writer
```

```python
Releasedate=soup.find('span',class_="sc-1thf9ly-2 dvggWt")
Releasedate.text
```

```python
Releasedate=[]

for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    Releasedate.append(i.text)
    
Releasedate
```
